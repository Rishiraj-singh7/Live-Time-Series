```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla=pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head()
```

```python
#simple moving avg
```

```python
df_tesla['Open'].plot(figsize=(12,6))
```

```python
df_tesla['Open:10 days rolling']=df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:10 days rolling']].plot(xlim=['2020-01-01','2020-01-01'],figsize=(15,6))
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(window=30,min_periods=1).mean()
df_tesla['Open:50 days rolling']=df_tesla['Open'].rolling(window=50,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(xlim=['2020-01-01','2020-01-01'],figsize=(15,7))
```

```python
#cummulative moving avg
#expnding
df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
#EMA Exponential Moving Average
#ema tesla share    smoothing figure - 0.1
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1','Open:50 days rolling']].plot(xlim=['2020-01-01','2020-01-01'],figsize=(15,7))
```

```python
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1','EMA_0.3','Open:50 days rolling']].plot(xlim=['2020-01-01','2020-01-01'],figsize=(15,7))
```

```python
#EwMA Exponential Moving Average
#ewma tesla share    smoothing figure - 0.1
df_tesla['EMA_5days'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open','EMA_5days','EMA_0.1']].plot(figsize=(18,8))
```
